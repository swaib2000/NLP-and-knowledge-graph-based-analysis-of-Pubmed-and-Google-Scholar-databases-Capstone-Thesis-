# Iteraction Type GeneToDisease


```python
import pandas as pd
```

```python
PD_data = pd.read_csv("PD_pubtator_28.csv")
```

```python
gene_sent = pd.read_csv("PD_Genes_Sen_28.csv")
chemical_sent = pd.read_csv("PD_Disease_Sentences_28.csv")
```

```python
chemical_sent.shape
```

```python
chemical_sent.head(4)
```

```python
gene_sent.head(4)
```

```python
gene_sent.shape
```

```python
PD_data.shape
```

```python
merged_df = gene_sent.merge(chemical_sent, on=[ 'Sentences','PMID'])
```

```python
merged_df
```

```python
merged_df['Sentences'][0]
```

```python
merged_df.shape
```

```python
merged_df.head(4)
```

```python
# Sample sentence and list of lowercase words
sentence = "This is a SAmple sentence to check for words in a list."
word_list = ["sample"]
# Convert the sentence to lowercase and split it into words
words_in_sentence = sentence.lower().split()

# Convert the list of words to lowercase
word_list = [word.lower() for word in word_list]

# Initialize a boolean variable
found_in_list = any(word in word_list for word in words_in_sentence)

# Print the result as True or False
print(found_in_list)
```

```python
inhibition_synonyms = ["inhibition","suppression", "restraint", "repression", "constraint", "blockade", "hindrance", "constraint", "obstruction", "interference", "limitation"]
activation_synonyms = [
    "stimulation",
    "initiation",
    "triggering",
    "induction",
    "turn-on",
    "activation",
    "excitation",
    "enabling",
    "catalysis",
    "start",
    "commencement",
    "launch",
]
proliferation_synonyms = ["growth", "multiplication", "expansion", "reproduction", "propagation", "amplification", "proliferation", "cell division", "increase"]

allosteric_synonyms = ["allosteric", "regulatory", "modulatory", "cooperative", "regulation", "modulation", "cooperativity", "non-active site", "indirect", "allosterically"]
agonist_synonyms = ["activator", "stimulator", "inducer", "enhancer", "mimetic","agonist"]

antagonist_synonyms = ['antagonist','inhibitor', 'blocker', 'opposer', 'adversary', 'challenger', 'competitor', 'foe', 'enemy', 'rival']
```

```python
merged_df['Interaction Type'] = None
```

```python
for index, row in merged_df.iterrows():
  sentence =  row['Sentences']
  words_in_sentence = sentence.lower().split()
  if(any(word in inhibition_synonyms for word in words_in_sentence)):
    merged_df.at[index, 'Interaction Type'] = "inhibition"
  elif(any(word in activation_synonyms for word in words_in_sentence)):
    merged_df.at[index, 'Interaction Type'] = "activation"
  elif(any(word in proliferation_synonyms for word in words_in_sentence)):
    merged_df.at[index, 'Interaction Type'] = "proliferation"
  elif(any(word in allosteric_synonyms for word in words_in_sentence)):
    merged_df.at[index, 'Interaction Type'] = "allosteric"
  elif(any(word in agonist_synonyms for word in words_in_sentence)):
    merged_df.at[index, 'Interaction Type'] = "agonist"
  elif(any(word in antagonist_synonyms for word in words_in_sentence)):
    merged_df.at[index, 'Interaction Type'] = "antagonist"
  else:
    merged_df.at[index, 'Interaction Type'] = "other"

```

```python
merged_df['Interaction Type'].value_counts()
```

```python
merged_df
```

```python
merged_df = merged_df.drop('PMCID_x', axis=1)
```

```python
merged_df
```

```python
up_words = ['up', 'above', 'skyward', 'elevated', 'ascendant', 'higher', 'upward']
down_words = ["lower", "beneath", "under", "beneath", "beneath", "beneath", "beneath", "beneath", "beneath", "beneath"]
up_regulated_words = ["up regulated", "upregulated", "elevated", "increased", "enhanced"]
down_regulated_terms = ['downregulated', 'reduced', 'decreased', 'lowered', 'diminished', 'suppressed']
```

```python
merged_df['Regulation'] = None
```

```python
for index, row in merged_df.iterrows():
  sentence =  row['Sentences']
  words_in_sentence = sentence.lower().split()
  if(any(word in up_words for word in words_in_sentence)):
    merged_df.at[index, 'Regulation'] = "up"
  elif(any(word in down_words for word in words_in_sentence)):
    merged_df.at[index, 'Regulation'] = "down"
  elif(any(word in up_regulated_words for word in words_in_sentence)):
    merged_df.at[index, 'Regulation'] = "up regulated"
  elif(any(word in down_regulated_terms for word in words_in_sentence)):
    merged_df.at[index, 'Regulation'] = "down regulated"
  else:
    merged_df.at[index, 'Regulation'] = "other"

```

```python
merged_df.to_csv("PD_Interaction_GeneToDisease_28.csv", index=False)
```
